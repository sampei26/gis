```python
import geopandas as gpd
import pandas as pd
```

```python
iwate_ku = gpd.read_file('N03-23_03_230101.geojson')
```

```python
iwate_pop = pd.read_csv('iwate.csv')
```

```python
iwate_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
iwate_ku
```

```python
iwate_ku['KEY_CODE'] = iwate_ku['KEY_CODE'].astype(int)
```

```python
iwate_ku_pop = iwate_ku.merge(iwate_pop,on='KEY_CODE')
```

```python
iwate_ku_pop.info()
```

```python
iwate_ku_pop['人口総数'] = iwate_ku_pop['人口総数'].astype(int)
```

```python
iwate_ku_pop.plot(column='人口総数',legend=True)
```

面白かったです。
人口分布など地域ごとの情報をマップ等可視化できる形で図示できれば今までにない視点から様々な課題を発見できそれが地域創成につながるのではないかと考えました。
